```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
dataset=load_breast_cancer()
```

```python
# Print the target variable
print("Target variables:",dataset['target_names'])
```

```python
# Count of the target variable
(unique,counts)=np.unique(dataset['target'],return_counts=True)
print("Unique values of the target variable:",unique)
print("Count of the target variable:",counts)
```

```python
sns.barplot(x=dataset['target_names'],y=counts)
plt.title("Target Variable Counts")
plt.show()
```

#Findings
* The Benign is the most common type of cancer because it has the highest value.

```python
# Create a margin of separating between the data points
X = dataset.data
y = dataset.target
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=dataset.feature_names)
df['target'] = y
```

```python
# Apply scaling to create clearer separation (margin)
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled_features = scaler.fit_transform(df[dataset.feature_names])
df_scaled = pd.DataFrame(scaled_features, columns=dataset.feature_names)
df_scaled['target'] = y
```

```python
# Plot two features that separate well
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df_scaled['mean radius'],
    y=df_scaled['mean texture'],
    hue=df_scaled['target'],
    palette=['red', 'blue']
)
plt.title("Breast Cancer Dataset (Scaled for Separation)")
plt.xlabel("Mean Radius (scaled)")
plt.ylabel("Mean Texture (scaled)")
plt.legend(labels=dataset.target_names)
plt.show()
```
